**LOGISTIC** **REGRESSION**

```python
class LogisticRegression():

  def __init__(self, learning_rate, no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations


  def fit(self,X ,Y ):

    self.m, self.n = X.shape
    self.X = X
    self.Y = Y

    self.w = np.zeros(self.n)
    self.b = 0

    for i in range(self.no_of_iterations):
        self.update_weights()


  def update_weights(self):

    Y_hat = 1 / (1 + np.exp( - (self.X.dot(self.w) + self.b)))

    dw = (1 / self.m) * np.dot(self.X.T, (Y_hat - self.Y))
    db = (1 / self.m) * np.sum(Y_hat - self.Y)

    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db

  def predict(self, X):

    Y_pred = 1 / (1 + np.exp( - (X.dot(self.w) + self.b)))
    Y_pred = np.where(Y_pred > 0.5, 1, 0) # INSTEAD OF IF CONDITION
    return Y_pred
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=2)
```

```python
classifier = LogisticRegression(learning_rate=0.01,no_of_iterations=1000)
classifier.fit(X_train, Y_train)
```

```python
train = classifier.predict(X_train)
train_accuracy = accuracy_score(Y_train, train)
print(train_accuracy)
```

```python
test = classifier.predict(X_test)
test_accuracy = accuracy_score(Y_test, test)
print(test_accuracy)
```

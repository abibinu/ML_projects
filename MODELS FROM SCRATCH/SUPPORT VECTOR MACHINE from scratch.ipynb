IMPORTING LIBRARIES

```python
import numpy as np
```

**SUPPORT VECTOR MACHINE**

```python
class SVM_classifier():

  def __init__(self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter

  def fit(self, X, Y):

    self.m, self.n = X.shape

    self.X = X
    self.Y = Y

    self.w = np.zeros(self.n)
    self.b = 0

    for i in range(self.no_of_iterations):

      self.update_weights()

  def update_weights(self):

      # LABEL ENCODING(1,-1)
      y_label = np.where(self.Y <= 0, -1, 1)

      # GRADIENTS(dw,db)
      for index,x_i in enumerate(self.X): # enumerate is used to print the index value also

        condition = (y_label[index] * (np.dot(self.X[index],self.w) - self.b)) >= 1

        if condition:

          dw = 2 * self.lambda_parameter * self.w
          db = 0

        else:

          dw = 2 * self.lambda_parameter * self.w - np.dot(x_i,y_label[index])
          db = y_label[index]


        # UPDATING THE PARAMETERS
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db

  def predict(self, X):

      output = np.dot(X,self.w) + self.b
      predicted_labels = np.sign(output)

      return np.where(predicted_labels <= -1, 0, 1)

```

IMPLEMENTATION

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

DATA SPLITTING

```python
X = diabetesdf.drop(columns = 'Outcome', axis = 1)
Y = diabetesdf.Outcome
```

```python
print(X)
print(Y)
```

DATA PREPROCESSING

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
print(X)
```

TEST TRAIN SPLIT

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2, stratify = Y)
```

TRAINING THE DATA

```python
classifier = SVM_classifier(learning_rate = 0.01, no_of_iterations = 1000, lambda_parameter = 0.001)
classifier.fit(X_train,Y_train)
```

```python
train_pred = classifier.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_pred)
print(train_accuracy)
```

```python
test_pred = classifier.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred)
print(test_accuracy)
```

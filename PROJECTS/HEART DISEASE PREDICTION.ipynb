IMPORTING DEPENDENCIES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND PROCESSING

```python
heart_data = pd.read_csv('/content/drive/MyDrive/ML Datasets/heart.csv')
```

```python
heart_data.head()
```

```python
heart_data['target'].value_counts()
```

SEPERATING FEATURES AND LABELS

```python
X = heart_data.drop(columns = 'target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
print(Y)
```

SPLITTING INTO TRAIN AND TEST DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,stratify = Y, test_size = 0.1, random_state = 2)
```

```python
X_train.shape
```

LOGISTIC REGRESSION MODEL TRAINING

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

PREDICTION AND ACCURACY SCORE TESTING

```python
pred = model.predict(X_train)
```

```python
acc = accuracy_score(pred,Y_train)
print(acc)
```

```python
result = model.predict(X_test)
```

```python
acc2 = accuracy_score(result,Y_test)
print(acc2)
```

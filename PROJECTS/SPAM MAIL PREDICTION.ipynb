Importing Dependencies

```python
# Numpy for making arrays
import numpy as np

# Pandas for making Dataframes
import pandas as pd

# To split data into Train and Test
from sklearn.model_selection import train_test_split

# Vectorizer for converting Text into Numerical values
# Because ML model cannot predict using Text values
from sklearn.feature_extraction.text import TfidfVectorizer

# Importing Logistic Regression Model
from sklearn.linear_model import LogisticRegression

# Importing Accuracy Score checker
from sklearn.metrics import accuracy_score
```

Data Collection and Pre-processing

```python
# Loading the data from csv file to pandas DF
raw_mail_data = pd.read_csv('/content/drive/MyDrive/ML Datasets/mail_data.csv')
```

```python
print(raw_mail_data)
```

Replace all missing values (if exists) with null String

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

Label Encoding

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
mail_data.head()
```

Seperating text and labels

```python
X = mail_data['Message']
Y = mail_data['Category']
```

Splitting into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 3)
```

```python
X_train.shape
```

Feature Extraction

```python
# Transform text data to feature vectors (numerical values)
# Loading tfidf vectorizer
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)
```

```python
# Transform X_test and X_train to vectors
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
# Convert labels into Int datatypes
Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)
```

```python
print(Y_train)
```

```python
print(X_train_features)
```

Training the Logistic Regression model

```python
# loading the model
model = LogisticRegression()

# Training the model
model.fit(X_train_features,Y_train)
```

```python
# Predicting using training data
X_train_pred = model.predict(X_train_features)
```

```python
# Accuracy score check
acc_train = accuracy_score(X_train_pred,Y_train)
```

```python
print("ACCURACY SCORE OF PREDICTION USING TRAINING DATA IS ", acc_train)
```

```python
# Predicting using Test data
result = model.predict(X_test_features)
```

```python
# checking the accuracy score
acc_result = accuracy_score(result,Y_test)
```

```python
print("ACCURACY SCORE OF PREDICTION USING TEST DATA IS ", acc_result)
```
